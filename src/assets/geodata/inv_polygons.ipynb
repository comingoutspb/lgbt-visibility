```python
def invert_polygon_coordinates(feature):
    """
    Inverts the order of coordinates for each ring in the polygon of the given GeoJSON feature.
    Assumes the feature's geometry type is either 'Polygon' or 'MultiPolygon'.
    """
    geom_type = feature['geometry']['type']
    if geom_type == 'Polygon':
        feature['geometry']['coordinates'] = [
            ring[::-1] for ring in feature['geometry']['coordinates']
        ]
    elif geom_type == 'MultiPolygon':
        feature['geometry']['coordinates'] = [
            [[ring[::-1] for ring in polygon] for polygon in multi_polygon]
            for multi_polygon in feature['geometry']['coordinates']
        ]

def invert_selected_geojson_polygons(json_data, feature_ids):
    """
    Inverts the polygons for selected features based on their IDs in the given GeoJSON object.
    """
    for feature in json_data['features']:
        if feature['id'] in feature_ids and feature['geometry']['type'] in ['Polygon', 'MultiPolygon']:
            invert_polygon_coordinates(feature)


# Function to save JSON data to a file
def save_json_file(data, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
        
```
